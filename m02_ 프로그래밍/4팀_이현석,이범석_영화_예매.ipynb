```python
import datetime
import time
import pytz

# 영화 제목 및 가격 딕셔너리
movie_list = {
    1 : ("혹성탈출" , 12000),
    2 : ("범죄도시" , 8000),
    3 : ("스턴트맨" , 15000),
    4 : ("쿵푸팬더" , 12000),
    5 : ("파묘" , 10000),
}


# 영화 제목에 따른 시간 리스트
movie_time1 = ["9시","10시 30분","11시","12시 20분","13시 50분","14시 30분"]
movie_time2 = ["9시 30분","10시 10분","11시","12시 35분","13시 10분","14시"]
movie_time3 = ["8시","8시 30분","10시 10분","12시 30분","13시"]
movie_time4 = ["7시","8시 30분","10시","14시","15시 30분"]
movie_time5 = ["9시","10시","11시","12시"]

# 영화 제목에 따른 좌석 리스트 _ 1
movie_sit_list = ['1', '2', '3', '4', '5']

# 영화 제목에 따른 좌석 리스트 _ 2
movie_sit_list_a = ['A1', 'A2', 'A3', 'A4', 'A5']
movie_sit_list_b = ['B1', 'B2', 'B3', 'B4', 'B5']
movie_sit_list_c = ['C1', 'C2', 'C3', 'C4', 'C5']
movie_sit_list_d = ['D1', 'D2', 'D3', 'D4', 'D5']
movie_sit_list_e = ['E1', 'E2', 'E3', 'E4', 'E5']
movie_sit_list_f = ['F1', 'F2', 'F3', 'F4', 'F5']
movie_sit_list_g = ['G1', 'G2', 'G3', 'G4', 'G5']
movie_sit_list_h = ['H1', 'H2', 'H3', 'H4', 'H5']

# 매점 메뉴 및 가격 딕셔너리
menu_items = {
    '1': ("Cheeze Nacho", 5000),
    '2': ("Hotdog", 3500),
    '3': ("Popcorn", 6000),
    '4': ("Squid", 10000)
}

# 추가 메뉴 가격
other_price = 1000

# odd_order (장바구니)에 담기는 딕셔너리
orders = {}

# 전체 금액에 대한 초기값
price_total = 0

class movie:

  def display_movie(self):
    print("영화관에 오신 것을 환영합니다 ♣♣♣")
    print("========================================")
    print("-        영화 제목 및 가격 -")
    print()
    for key, (name, price) in movie_list.items():
        print(f"  {key}. {name} : {price}원")
    print()
    tz = pytz.timezone('Asia/Seoul')
    now = datetime.datetime.now(tz)
    now_time = now.strftime('%H시 %M분')

    print(f"현재 시간은 {now_time}입니다")
    print("========================================")

  def diply_time(self):
    what = input("무슨 영화를 관람하시겠습니까 >> ")
    if what == '혹성탈출':
      print(movie_time1)
    elif what == '범죄도시':
      print(movie_time2)
    elif what == '스턴트맨':
      print(movie_time3)
    elif what == '쿵푸팬더':
      print(movie_time4)
    elif what == '파묘':
      print(movie_time5)
    else:
      print("죄송하지만 상영하지 않는 영화입니다")
    while True:
        time = input("시간을 골라 주세요 >> ")
        if what == '혹성탈출' and time in movie_time1:
            break
        elif what == '범죄도시' and time in movie_time2:
            break
        elif what == '스턴트맨' and time in movie_time3:
            break
        elif what == '쿵푸팬더' and time in movie_time4:
            break
        elif what == '파묘' and time in movie_time5:
            break
        else:
            print("그 시간에 상영하지 않습니다 ")
    person = int(input("몇 명이서 오셨나요? >> "))
    print()
    print("----------- 좌석표 -----------")
    print()
    print("=========== 스크린 ===========")
    print(movie_sit_list_a)
    print(movie_sit_list_b)
    print(movie_sit_list_c)
    print(movie_sit_list_d)
    print(movie_sit_list_e)
    print(movie_sit_list_f)
    print(movie_sit_list_g)
    print(movie_sit_list_h)
    print()
    num1 = input("관람하실 열을 골라주세요 (A~H) >> ").upper()

    # 입력이 A~H 사이의 알파벳이 아닌 경우를 체크
    if num1 not in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']:
       return print("잘못된 섹션입니다.")

    # 해당 알파벳 열에 대응하는 좌석 리스트를 가져옴
    seat_list = globals().get(f'movie_sit_list_{num1.lower()}')

    # 해당 열의 사용 가능한 좌석을 출력함.
    print(f"{num1}의 사용 가능한 좌석:")
    print(seat_list)

    # 빈 리스트를 생성하여 예약한 좌석의 번호를 저장할 number 리스트를 초기화함.
    number = []

    # 반복문을 통해 person 값만큼 좌석을 예약한다.
    i = 1
    while i <= person:
        select = int(input(f"{i}번째 좌석을 선택하세요 (1-5번) >> "))

        # 선택한 좌석이 유효한 범위인지 확인함.
        if 1 <= select <= 5:

          # 선택한 좌석이 이미 예약되었는지 확인함.
          if seat_list[select - 1] == "●":
            print("이미 예약된 좌석이니, 다른 좌석을 선택하세요")
          else:

            # 좌석을 예약 처리하고 예약한 좌석 번호를 출력합니다.
            seat_list[select - 1] = "●"
            print(f"{select}번 좌석이 예약되었습니다")
            print("현재 좌석:", seat_list, "\n")
            number.append(select)
            i += 1
    else:
        # 예약이 모두 완료되면 예약한 좌석과 시간을 출력하고 결제 수단을 입력받습니다.
        print(f"예약한 좌석: {num1}열", tuple(number),"시간 : ",time)
        pay = input("결제하실 수단을 선택해주세요. (현금 / 카드)")
        while True:
              if pay == '현금':
                  for i in range(0, 5):
                        # 사용자가 입력 받은 영화 제목과 movie_list 딕셔너리에서 데이터를 가져옴. ex ) what : 혹성탈출 / movie_list[키값][0번째 밸류값]과 일치하는 지에 대한 여부 확인
                        if what == movie_list[i+1][0]:
                          Total_movie_price_1 = int(movie_list[i+1][1]) * person
                          print(f"영화 표 가격 : {Total_movie_price_1}원 입니다.")

                          cash_pay = int(input("현금을 넣어주세요. : "))
                          change = cash_pay - Total_movie_price_1
                          if cash_pay >= Total_movie_price_1:
                              print(f"{cash_pay}원 받았습니다. 총 금액 : {Total_movie_price_1}원이며, 거스름돈은 {change}원 입니다.")
                              break
                          # 금액이 적은 경우 해당 멘트 발송 후 금액
                          else:
                              print("금액이 맞지 않습니다. 다시 현금을 넣어주세요.")
              # pay : 현금이 아닌 경우
              elif pay == '카드':
                  Total_movie_price_2 = int(movie_list[i+1][1]) * person
                  print(f'카드로 결제 시도 중입니다.. \n● 결제금액은 {Total_movie_price_2}원 입니다.\n')

              else:
                  print("다시 입력해주세요. ")
              break
        input("결제를 완료하려면 아무 키나 누르세요...")  # 사용자가 결제를 확인할 수 있도록 입력 대기
        print(f"\n총 금액 {movie_list[i+1][1]}원이 결제 되었습니다. 감사합니다.")


# menu_items = {
#     '1': ("Cheeze Nacho", 5000),
#     '2': ("Hotdog", 3500),
#     '3': ("Popcorn", 6000),
#     '4': ("Squid", 10000)
# }

def display_menu():
    print("\n─--─━━ 매점에 오신 것을 환영합니다 ━━─--─")
    print("================================================")
    print("           - S n a c k M e n u -")
    print()
    for key, (name, price) in menu_items.items():
        print(f"  {key}. {name} : {price}원")
    print()
    print("   ★ 메뉴 추가 선택 시 탄산수를 추가로 드립니다. (가격 : 1,000원)")
    print()
    print("================================================")
    print()

def display_order_summary_1():
    print("\n  ■■ 장바구니 ■■")
    global price_total
    price_total = 0
    for key, (name, price) in menu_items.items():
        count = orders.get((key, False), 0)
        count_set = orders.get((key, True), 0)
        print(f" {name} {count}개: {count * price}원")
        print(f" {name} 세트 {count_set}개: {count_set * (price + other_price)}원")
        price_total += (count * price) + (count_set * price)
    print(f" 총 금액 {price_total}원\n")

def add_order(menu_id, is_set):
    item_name, item_price = menu_items[menu_id]
    if is_set:
        item_price += other_price
    num = input(f"\n● {item_name}{' 세트' if is_set else ''} 몇 개 주문 하시겠습니까?\n")
    if num.isdigit() and int(num) > 0:
        orders[(menu_id, is_set)] = orders.get((menu_id, is_set), 0) + int(num)
        print(f"\n{item_name}{' 세트' if is_set else ''} {num}개가 장바구니에 담겼습니다.\n")
    else:
        print('★양수만 입력하세요.★')

def process_order():
    while True:
        # 사용자 입력을 받아 strip() 함수를 사용하여 양쪽 공백을 제거 / 소문자로 변환
        order = input("● 주문을 하시겠습니까? (y 또는 n) >> ").strip().lower()
        if order == 'y':
          while True:
            display_order_summary_1()
            choice = input("● 주문할 메뉴를 골라주세요. ( 1: 'Cheeze Nacho' , 2 : 'Hotdog', 3 : 'Popcorn', 4 : 'Squid' )\
            전체 취소:0, 결재:5) >> ").strip().lower()
            if choice in menu_items:
                set_choice = input("\n● 세트 메뉴 주문하시겠습니까? ( 일반 주문 시 : n, 세트 주문 시 : y )\n").strip().lower()
                if set_choice in ['y', 'n']:
                   add_order(choice, set_choice == 'y')
                else:
                  print('올바르게 입력해주세요. (y 또는 n).')
            elif choice == '0':
                orders.clear()
                print('\n모든 주문이 취소되었습니다.')
            elif choice == '5':
                if price_total > 0:
                        pay = input("결제하실 수단을 선택해주세요. (현금 / 카드) : ")
                        while True:
                            if pay == '현금':
                                print(f'현금을 넣어주세요.. \n● 결제금액은 {price_total}원 입니다.\n')
                                cash_pay = int(input("현금을 넣어주세요. : "))
                                change = cash_pay - price_total
                                if cash_pay >= price_total:
                                    print(f"{cash_pay}원 받았습니다. 총 금액 : {price_total}원이며, 거스름돈은 {change}원 입니다.")
                                    break
                                else:
                                    print("금액이 맞지 않습니다. 다시 현금을 넣어주세요.")
                            else:
                                print(f'카드로 결제 시도 중입니다.. \n● 결제금액은 {price_total} 원 입니다.\n')
                                break
                        input("결제를 완료하려면 아무 키나 누르세요...")  # 사용자가 결제를 확인할 수 있도록 입력 대기
                        print(f"\n총 금액 {price_total}원이 결제 되었습니다. 감사합니다.")
                        orders.clear()
                        break
                else:
                    print('메뉴를 고르고 주문해주세요!')
            else:
                print('메뉴의 지정된 숫자로 입력하세요')
        else:
          print('● 이용해주셔서 감사합니다.')
          break




# 클래스 movie를 정의하고 그 안에 display_movie와 display_time 메서드를 호출한 후
m = movie()
m.display_movie()
m.diply_time()

# 매점 이용에 대한 유무 확인
cho = input("매점을 이용하시겠습니까 >> Y or N 로 입력하세요 >> ")
if cho == 'Y' or cho =='y':
   display_menu()
   process_order()
elif cho == 'N' or cho == 'n':
    print('● 이용해주셔서 감사합니다.')
```
