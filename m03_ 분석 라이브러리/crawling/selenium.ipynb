```python
# Inplicit Waits(암묵적 대기) : 찾으려는 element가 로드될 때까지 지정한 시간만큼 대기할 수 있도록 설정
# By 클래스는 Selenium에서 HTML 요소를 찾기 위한 다양한 방법을 제공
# webDriverWait 클래스는 특정 조건이 충족될 때까지 대기하는 기능
# expected_conditions 모듈은 여러가지 조건을 미리 정의하여 특정 요소가 나타나거나 특정 상태가 될 떄가지 기다릴 수 있도록 함
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get(url='http://www.google.com/')
try:
    element = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'gLFyf'))
    )
    # 요소가 존재하면 텍스트 출력 (이 경우에는 입력창이므로 text가 없을 수 있음)
    print("Element found : ",element.get_attribute('title'))
except Exception as e:
    # 예외 발생 시 메세지 출력
    print("An error occurred : ", e)
finally:
    # WebDriver 종료
    driver.quit()
```

```python
import time
# xpath 속성을 사용하여 접근하여 요소를 추출
try:
    xpath = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
    print("XPath:", xpath.get_attribute('title'))
except:
    print("Element not found by XPath:")

# 2. 앵커태그(a 태그)에 사용되는 텍스트로 접근하여 요소를 추출 (Gmail 링크)
try:
    gmail_link_text = driver.find_element(By.LINK_TEXT, "Gmail")
    print("Link Text:",gmail_link_text.text)
except:
    print("Element not found by Link Text:")

# 3. 앵커태그(a 태그)에 사용되는 일부 텍스트로 접근하여 요소를 추출 (이미지 링크)
try:
    partial_link_text = driver.find_element(By.PARTIAL_LINK_TEXT, "Gm")
    print("Partial Link Text:", partial_link_text.text)
except:
    print("Element not found by Partial Link Text:")

# 4. 태그를 사용하여 접근하여 요소를 추출 (첫 번째 a 태그)
try:
    first_anchor_tag = driver.find_element(By.TAG_NAME, "input")
    print("Tag Name:", first_anchor_tag.get_attribute('value'))
except:
    print("Element not found by Tag Name:")

# 5. CSS 선택자를 사용하여 접근하여 요소를 추출 (검색 버튼)
try:
    search_button_css = driver.find_element(By.CSS_SELECTOR, "input.gNO89b")
    print("CSS Selector:", search_button_css.get_attribute('value'))
except:
    print("Element not found by CSS Selector:")

time.sleep(5)
driver.quit()
```

1. getElementById()
 - 파라미터로 찾으려는 id를 전달하여, 해당 element를 찾을 수 있습니다. id는 유일한 값이므로, 하나의 element만 리턴합니다.
2. getElementsByClassName()
 - 클래스 이름으로 element를 찾아서,이 클래스 이름을 가지는 모든 element 목록을 리턴합니다.이 함수의 이름을 자세히 보면 getElementsByClassName으로 Element's'가 복수 형태인 것은 이 함수가 목록을 리턴하기 때문입니다.
3. getElementByTagName()
 - 위 코드는 'div' 태그를 가지는 모든 element 목록을 찾아줍니다.이 함수 역시, 다수의 element를 리턴하기 때문에 함수명에 복수형의 'elements'가 포함되어 있습니다.
4. querySelector()
 - DOM에서 원하는 element를 찾기 위해서 querySelector() 를 사용할 수도 있는데, 이 함수는 파라미터로 입력받은 CSS선택자를 사용해서, element를 찾아줍니다.
 - querySelector() 함수는, 파라미터로 입력받은 CSS 선택자로 찾은 여러개의 element 중 첫번째 element를 리턴합니다.
 - 태그 이름으로 element를 찾을 때는 태그명을 문자열로 넘겨줍니다.위 예제는 div태그를 가지는 element 중 첫번째 element를 리턴합니다.
5. querySelectorAll()
 - 사용법은 querySelector() 와 같습니다. 다만, querySelectorAll()은 이름에서 알수 있듯이, CSS선택자(파라미터로 넘겨준)로 찾은 모든 element 목록을 리턴합니다.

Selenium only vs with JavaScript
- Selenium만 사용하는 경우: 코드가 더 직관적이고 Python만으로 작성되며, 디버깅과 유지보수가 용이합니다. 하지만, 속도가 느릴 수 있습니다.

```python


from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# 웹 드라이버 설정
driver = webdriver.Chrome()

# HTML 파일 열기
driver.get("file:///path/to/file_task1.html")

# 예제 1: document.getElementById(id)
script_1 = "return document.getElementById('title').innerText;"
result_1 = driver.execute_script(script_1)
print(f"Title Text: {result_1}")

# 예제 2: document.getElementsByClassName(className)
script_2 = """
var elements = document.getElementsByClassName('content');
if(elements.length > 0) {
    return elements[0].innerHTML;
}
return null;
"""
result_2 = driver.execute_script(script_2)
print(f"Content Inner HTML: {result_2}")

# 예제 3: document.getElementsByTagName(tagName)
script_3 = """
var paragraphs = document.getElementsByTagName('p');
if(paragraphs.length > 0) {
    return paragraphs[0].innerText;
}
return null;
"""
result_3 = driver.execute_script(script_3)
print(f"Paragraph Text: {result_3}")

# 예제 4: document.querySelector(selector)
script_4 = "return document.querySelector('button#changeTextButton').innerText;"
result_4 = driver.execute_script(script_4)
print(f"Button Text: {result_4}")

# 예제 5: document.querySelectorAll(selector)
script_5 = """
var inputs = document.querySelectorAll('input[type=text]');
var values = [];
inputs.forEach(function(input) {
    values.push(input.value);
});
return values;
"""
result_5 = driver.execute_script(script_5)
print(f"Input Values: {result_5}")

# 브라우저 종료
driver.quit()
```

Q. 검색 키워드와 뉴스 기사 수를 입력한 후 다음 사항을 수행하는 크롤링 프로그램을 작성하세요.
- 크롤링한 내용에서 타이틀과 링크를 추출 
- 추출한 데이터를 판다스 데이터 프레임으로 변환하고 csv 파일로 저장
- csv 파일을 다시 불러와서 데이터 프레임으로 출력 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import pandas as pd
import os

# 사용자로부터 검색 키워드와 필요한 뉴스 기사 수 입력 받기
query = input('검색 키워드를 입력하세요 : ')
news_num = int(input('총 필여한 뉴스기사 수를 입력해주세여(숫자만 입력) : '))

# 네이버 뉴스 검색 URL
news_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'

# Selenium Webdrive 설정
driver = webdriver.Chrome()
driver.get(news_url.format(query))

SCROLL_PAUSE_TIME = 2
news_dict = {}
idx = 0
last_height = driver.execute_script("return document.body.scrollHeight")
print('크롤링 중....')

while idx < news_num:
    # 페이지의 HTML 소스를 가져와 BeautifulSoup으로 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    itmes = soup.select('.news_tit')
    for item in itmes:
        if idx >= news_num:
            break
        title = item.get_text()
        url = item.get('href')
        news_dict[idx] = {'title' : title, 'url': url}
        idx += 1 
    if idx >= news_num:
        break
    # 페이지를 끝까지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # 새로운 높이가 로드될 때가지 대기
    time.sleep(SCROLL_PAUSE_TIME)
    new_height =  driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break # 더 이상 스크롤할 내용이 없으면 종료
    last_height = new_height

driver.quit()
print('크롤링 완료')
# 데이터 프레임 변환
print('데이터프레임 변환')
news_df = pd.DataFrame(news_dict).T
# 결과 출력

folder_path = os.getcwd()
file_name = f'네이버뉴스_{query}.csv'
news_df.to_csv(file_name,index=False)
print(f'파일 저장 완료 | 경로 : {folder_path}\\{file_name}')
df = pd.read_csv(file_name)
df
```

html.fromstring(page_source) vs BeautifulSoup(page_source, 'lxml')
- 둘 다 HTML 파싱을 위한 방법이지만, 각각 다른 라이브러리를 사용.

html.fromstring(page_source) 방법
- 라이브러리: lxml
- 주요 기능:
    - XML과 HTML을 빠르고 효율적으로 파싱합니다.
    - XPath와 XSLT 지원이 강력합니다.
- 사용법:
    - XPath를 사용하여 요소를 선택합니다.
    - 성능이 뛰어나고 대규모 HTML 문서를 처리하는 데 적합합니다.

BeautifulSoup(page_source, 'lxml') 방법
- 라이브러리: BeautifulSoup
- 주요 기능:
    - HTML과 XML을 유연하게 파싱합니다.
    - 다양한 파서(lxml, html.parser 등)를 지원합니다.
    - HTML 구조가 비정상적이거나 잘못된 경우에도 잘 처리합니다.
- 사용법:
    - CSS 선택자 또는 BeautifulSoup 고유의 API를 사용하여 요소를 선택합니다.
    - 사용하기 쉽고 직관적입니다.

```python
import requests
from lxml import html

url = "https://web.joongna.com/search/%EC%95%84%EC%9D%B4%ED%8F%B0x?page=11"
response = requests.get(url)
tree = html.fromstring(response.content)

data = []

articles = tree.xpath('//*[@id="__next"]/div/main/div/div/ul/li')
for article in articles:
    try:
        title = article.xpath('.//div/div/a/div/h2/text()')[0]
        price = article.xpath('.//div/div/a/div/div/text()')[0]
        time_diff = article.xpath('.//div/div/a/div/div/span[last()]')[0].text_content()
        data.append({"title": title, "price": price, "Time Difference": time_diff})
    except IndexError:
        continue

data
```
