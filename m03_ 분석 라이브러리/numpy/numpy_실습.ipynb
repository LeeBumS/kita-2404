###  아래 가이드에 따라서 1000명의 고객 구매데이터를 생성하고 아래사항을 수행하세요(관측치 10000개)

#### Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.
고객별 속성 데이터
- id : 1 ~ 1000 고객 번호
- gender : 0,1 정수 난수 생성
- age : 0 ~ 80사이 정수 난수 생성
- region : 1 ~ 10 사이 정수 난수 생성

거래 데이터
- no : 1 ~ 10000번 일련번호
- day : 요일
- date : 2020-01-01 ~ 2021-12-31 사이의 날짜 1000개 생성
- product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- product_21 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- price_20 : 1000 ~ 50000 사이 실수 난수 생성
- price_21 : 1000 ~ 50000 사이 실수 난수 생성
- quantity_20 : 1 ~ 100 사이 정수 난수 생성
- quantity_21 : 1 ~ 100 사이 정수 난수 생성
- time_20 : 01 ~ 24 사이 정수 난수 생성
- time_21 : 01 ~ 24 사이 정수 난수 생성
- amount_20 : price_20 * quantity_21
- amount_21 : price_21 * quantity_21

#### Q2. df를 수정하여 5가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요

#### Q3. df 를 탐색적 분석을 통하여 인사이트를 도출하세요.

```python
#고객별 속성 데이터
import pandas as pd
import numpy as np

np.random.seed(1)

customer_id = np.arange(1, 1001)
gender = np.random.randint(0, 2, size=1000)
age = np.random.randint(1, 81, size=1000)
region = np.random.randint(1, 11, size=1000)

customer_df = pd.DataFrame({
    "id": customer_id,
    "gender": gender,
    "age": age,
    "region": region
})


df = customer_df
df.head()
```

```python
# 거래 데이터
import pandas as pd
import random
import datetime
import numpy as np

nos = np.arange(1, 10001)
customer_ids = np.random.choice(customer_id, size=10000)
days = np.random.choice(['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'], size = 10000)
dates = pd.date_range(start='2020-01-01', end='2021-12-31').to_series().sample(n = 10000, replace=True).values
product_20 = np.random.randint(10, 31, size = 10000)
product_21 = np.random.randint(10, 31, size = 10000)
price_20 = np.random.uniform(1000, 50000, size = 10000)
price_21 = np.random.uniform(1000, 50000, size = 10000)
quantity_20 = np.random.randint(1, 101, size = 10000)
quantity_21 = np.random.randint(1, 101, size = 10000)
time_20 = np.random.randint(1, 25, size = 10000)
time_21 = np.random.randint(1, 25, size = 10000)
amount_20 = price_20 * quantity_20
amount_21 = price_21 * quantity_21


transactions_df = pd.DataFrame({
    'no': nos,
    'customer_id': customer_ids,
    'day': days,
    'date': dates,
    'product_20': product_20,
    'product_21': product_21,
    'price_20': price_20,
    'price_21': price_21,
    'quantity_20': quantity_20,
    'quantity_21': quantity_21,
    'time_20': time_20,
    'time_21': time_21,
    'amount_20': amount_20,
    'amount_21': amount_21
})

transactions_df['price_20'] = round(transactions_df['price_20'], 2)
transactions_df['price_21'] = round(transactions_df['price_21'], 2)
transactions_df['amount_20'] = round(transactions_df['amount_20'], 2)
transactions_df['amount_21'] = round(transactions_df['amount_21'], 2)

df = pd.merge(transactions_df, customer_df, how='left', left_on='customer_id', right_on='id')

df.drop('id', axis=1, inplace=True)

df.head()

```

```python
df.info()
```

```python
# 요일별 거래량 계산
day_count = df[df['day'].isin(['월요일', '화요일', '수요일', '목요일', '금요일'])]['day'].count()
end_count = df[df['day'].isin(['토요일', '일요일'])]['day'].count()

print("평일 거래량 : ",day_count)
print("주말 거래량 : ",end_count)
```

```python

# 연도별 거래액 계산
amount_2020 = df[df['date'].dt.year == 2020]['amount_20'].sum()
amount_2021 = df[df['date'].dt.year == 2021]['amount_21'].sum()

# 연도별 거래 건수 계산
total_2020 = len(df[df['date'].dt.year == 2020])
total_2021 = len(df[df['date'].dt.year == 2021])

print('2020년도 거래액 : ', amount_2020)
print('2021년도 거래액 : ', amount_2021)
print('2020년도 거래수 : ', total_2020)
print('2021년도 거래수 : ', total_2021)
```

```python
# 고객 연령대별 구매
import pandas as pd

# 고객 연령대별로 데이터 그룹화
age_groups = df.groupby(pd.cut(df['age'], bins=[10, 20, 30, 40, 50, 60, 70, 80],
                               labels=['10대', '20대', '30대', '40대', '50대', '60대', '70대']))
print("2020년")
# 연령대별 구매,평균  계산 20년도
for age_group, group_data in age_groups:
    count_20 = len(group_data['amount_20'])
    amount_20 = group_data['amount_20'].mean()
    print(f"{age_group}: 20년도 구매건수 {count_20}, 평균거래액 {amount_20:.2f} 입니다")

print()
age_groups = df.groupby(pd.cut(df['age'], bins=[10, 20, 30, 40, 50, 60, 70, 80],
                               labels=['10대', '20대', '30대', '40대', '50대', '60대', '70대']))

# 21년도 구매건수, 평균 계산
print("2021년")
for age_group, group_data in age_groups:
    count_21 = len(group_data['amount_21'])
    amount_21 = group_data['amount_21'].mean()
    print(f"{age_group}: 21년도 구매건수 {count_21}, 평균거래액 {amount_21:.2f} 입니다")
```
