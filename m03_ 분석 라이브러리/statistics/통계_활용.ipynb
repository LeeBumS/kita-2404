Q. 한 학급의 학생 10명의 수학 시험 점수는 다음과 같습니다:

[85, 72, 90, 68, 88, 76, 95, 89, 77, 84].

학급의 평균 점수와 표준편차를 계산하여 전체 성적 수준과 변동성을 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import numpy as np

scores = np.array([85, 72, 90, 68, 88, 76, 95, 89, 77, 84])
mean_score = np.mean(scores)
print(f"Mean score: {mean_score}")

standard_deviation = np.std(scores)
print(f"Standard deviation: {standard_deviation}")

#평균 점수는 학급 전체의 성적이 평균 이상임을 나타냅니다.
#표준편차는 확인하니 일부 학생들은 평균보다 상당히 우수하거나 낮습니다.
```

Q. 어느 도시에 지난 1년간 매달 내린 강수량 데이터는 다음과 같습니다:

[85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95].

이 데이터를 이용해 중간값과 1사분위수, 3사분위수를 계산하여 강수량의 분포를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import numpy as np

rainfall_data = np.array([85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95])

# 중간값 계산
median = np.median(rainfall_data)

# 제1사분위수(Q1) 계산
q1 = np.percentile(rainfall_data, 25)

# 제3사분위수(Q3) 계산
q3 = np.percentile(rainfall_data, 75)

# 결과 출력
print("중간 강수량:", median)
print("제1사분위수(Q1):", q1)
print("제3사분위수(Q3):", q3)

# 1. 중간 강수량은 89mm로, 절반의 달은 89mm 미만의 강수량을 받고 나머지 절반은 89mm 이상의 강수량을 받았음을 나타냅니다.
# 2. 제1사분위수(Q1)은 80mm로, 25%의 달은 80mm 미만의 강수량을 받았음을 나타냅니다.
# 3. 제3사분위수(Q3)은 100mm로, 75%의 달은 100mm 미만의 강수량을 받았음을 나타냅니다.
# 4. Q1과 Q3 사이의 범위(사분위 범위)는 20mm로, 강수량 분포가 연중 내내 비교적 일관됨을 시사합니다.
```

Q. 한 온라인 쇼핑몰의 지난 10일간 일일 방문자 수는 다음과 같습니다:

[150, 200, 220, 180, 190, 170, 210, 230, 160, 175].

방문자 수의 분포를 시각화하기 위해 히스토그램을 그리고, 방문자 수의 변동성에 대한 인사이트를 도출하세요.

```python
import matplotlib.pyplot as plt
import pandas as pd

visitors = [150, 200, 220, 180, 190, 170, 210, 230, 160, 175]

plt.hist(visitors, bins=30, alpha=0.5)
plt.title('Distribution of Visitors')  # 방문자 수 분포
plt.xlabel('Number of Visitors')  # 방문자 수
plt.ylabel('Frequency')  # 빈도
plt.show()



# 평균과 표준편차 계산
mean = sum(visitors) / len(visitors)
std = (sum((x - mean)**2 for x in visitors) / len(visitors))**0.5

# 인사이트 출력
print("평균 방문자 수는", mean, "명입니다.")
print("표준편차는", std, "명입니다.")
print("이는 방문자 수가 평균", std, "명 정도로 변동")
```

Q. 한 식당의 월별 홍보 비용과 월별 방문자 수 데이터가 각각 다음과 같습니다:

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
# 홍보 비용과 방문자 수 데이터
promotion_cost = np.array([500, 600, 700, 800, 900])
visitor_count = np.array([300, 400, 500, 600, 700])

# 상관 계수 계산
correlation = np.corrcoef(promotion_cost, visitor_count)[0, 1]
print(f"Correlation coefficient: {correlation}")

# 인사이트 해석
if correlation > 0.7:
  print("홍보 비용과 방문자 수 간에 강한 양의 상관관계가 있습니다")
elif correlation < 0.7:
  print("홍보 비용과 방문자 수 간에 강한 음의 상관관계가 있습니다")
else:
  print("홍보 비용과 방문자 수 간에 상관관계 약하거나 없습니다")
```

Q. 한 지역의 주택 가격 데이터가 다음과 같습니다:

[250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000].

주택 가격의 분포를 시각화하기 위해 분포 그래프를 그리고, 주택 가격 분포에 대한 인사이트를 도출하세요.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 주택 가격 데이터
house_prices = np.array([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

# 팔도 그래프 그리기
sns.kdeplot(house_prices)
plt.title('Density Plot of House Price')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()

# 인사이트 해석
mean_price = np.mean(house_prices)
median_price = np.median(house_prices)

print(f"Mean Price: {mean_price}, Median Price: {median_price}")

if mean_price > median_price:
  print("주택 가격 분포가 오른쪽으로 치우쳐 있음을 의미합니다")
elif mean_price < median_price:
  print("주택 가격 분포가 왼쪽으로 치우쳐 있음을 의미합니다")
else:
  print("주택 가격이 고르게 분포되어 있음을 의미합니다")
```

Q. 한 교육기관의 시험 점수 데이터가 다음과 같습니다:

[55, 60, 65, 70, 75, 80, 85, 90, 95, 100].

시험 점수의 도수 분포표를 작성하고, 이를 통해 성적 분포에 대한 인사이트를 도출하세요.

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
rnd_investment = np.array([100, 200, 300, 400, 500])
annual_revenue = np.array([200, 400, 600, 800, 1000])

# 추세선이 있는 산전도 그리기
plt.scatter(rnd_investment, annual_revenue)
m, b = np.polyfit(rnd_investment, annual_revenue, 1)
plt.title("Scatter Plot with Trend Line")
plt.xlabel('R&D Investment (thousand dollars)')
plt.ylabel('Annual Revenue (thousand dollars)')
plt.show()

correlation = np.corrcoef(rnd_investment, annual_revenue)[0,1]
print(f"Correlation coefficient: {correlation}")

if correlation > 0.7:
  print("R&D 투자와 연간 수익 간에 강한 양의 상관관게가 있습니다")
else:
  print("R&D 투자와 연간 수익 간에 상관관게가 없습니다")
```
