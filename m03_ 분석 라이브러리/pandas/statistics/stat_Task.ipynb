```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
ar1 = np.random.normal(50, 10, 100)
ar1_median = np.median(ar1)
ar1_mean = np.mean(ar1)

ar2 = ar1.copy()
ar2[0] = 500
ar2_median = np.median(ar2)
ar2_mean = np.mean(ar2)

# 중위수와 평균 차이 출력
print(ar2_median - ar1_median)
print(ar2_mean - ar1_mean)
```

```python
# 시각화
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# ar1 분포 시각화 (히스토그램과 KDE)
sns.histplot(ar1, bins=30, kde=True, ax=axs[0])
axs[0].set_title('ar1 Distribution')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# ar2 분포 시각화 (히스토그램과 KDE)
sns.histplot(ar2, bins=30, kde=True, ax=axs[1])
axs[1].set_title('ar2 Distribution')
axs[1].set_xlabel('Value')

plt.tight_layout()
plt.show()
```
